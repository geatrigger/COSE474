```python
import os
from tqdm.notebook import tqdm

# path_dir = './darknet/data/obj/Palm tree/'
# path_dir = './darknet/data/test/Palm tree/'
# path_dir = './darknet/data/obj/Car/'
path_dir = './darknet/data/test/Car/'
# path_dir = './darknet/data/obj/Car/'
# path_dir = './darknet/data/test/Car/'
file_list = os.listdir(path_dir)
# print(file_list)
```

```python
for name in tqdm(file_list):
    if name[-4:] == '.txt':
        print(name)
        content = ''
        with open(path_dir + name, 'r') as f:
            content = f.read()
            lines = content.split('\n')
#             print(lines)
        for sIndex in range(0, len(lines)):
            if len(lines[sIndex]) > 0:
                lines[sIndex] = '0' + lines[sIndex][1:]
        print('\n'.join(lines))
        with open(path_dir + name, 'w') as f:
            f.write('\n'.join(lines))
```
